```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

## Read Images

```python
my_image = cv2.imread('/content/drive/MyDrive/Images/f2.jpeg')
```

## Display with meplotlib

## Change Color Space To RBG

```python
bgr_img = cv2.cvtColor(my_image, cv2.COLOR_BGR2RGB)
print(bgr_img.shape)
plt.imshow(bgr_img)
```

## Save Image

```python
cv2.imwrite('bgr_img.png', bgr_img)
```

## Load Gray Scale Image

```python
gray_img = cv2.imread('/content/drive/MyDrive/Images/pexels-rodolfoclix-922610.jpg', cv2.IMREAD_GRAYSCALE)
resized_img = cv2.resize(gray_img, (300, 300))
cv2_imshow(resized_img)
```

```python
plt.figure(figsize=(5,5))
plt.imshow(gray_img, cmap='gray') # added cmap='gray'
```
